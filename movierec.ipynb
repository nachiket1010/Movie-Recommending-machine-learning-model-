```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
    return L
```

```python
import pandas as pd

# Assuming 'movies' is your DataFrame
# Define a custom function to handle None values
def handle_none_values(x):
    return x if x is not None else []

# Apply the custom function to handle None values in the 'keywords' column
movies['keywords'] = movies['keywords'].apply(handle_none_values)

# Now, apply the lambda function to remove spaces from the keywords
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head(25)
```

```python
movies['tags']=movies['overview']+movies['genres']+ movies['keywords']
```

```python
movies.head(10)
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df
```

```python
import pandas as pd

# Assuming 'new_df' is your DataFrame
# Define a custom function to convert all elements to strings
def convert_to_string(x):
    if isinstance(x, list):
        return [str(item) for item in x]
    else:
        return [str(x)]

# Convert all elements in the 'tags' column to strings
new_df['tags'] = new_df['tags'].apply(convert_to_string)

# Now, apply the lambda function to join the elements in the 'tags' column
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags']
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
 new_df['tags'][1]
```

```python
#bagofwords
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python

#stemming
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem (text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
ps.stem('loving')
```

```python
new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend (movie):
  movie_index = new_df[new_df['title']==movie].index[0]
  distances= similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
